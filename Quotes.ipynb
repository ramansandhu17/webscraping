```python
#import libraries

from bs4 import BeautifulSoup as BS
import requests
from collections import Counter

URL="http://quotes.toscrape.com/"

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}

response = requests.get(URL,headers=headers)

soup1=BS(response.text,"html.parser")

soup2=BS(soup1.prettify(),"html.parser")

#print(soup2)


#lets get the Authors information 

authors = soup2.find_all("small",class_="author")

authors_elements = [authors_elements.text.strip() for authors_elements in authors]
print("Authors:",authors_elements)

sorted_authors=sorted(authors_elements)
print("The Authors sorted are:",sorted_authors)


# we want to check the tags that are used mostly and the number of times they are used

count=soup2.find_all("a",class_="tag")

tags=[tags.text.strip() for tags in count]

tags_count=Counter(tags) #this counts the times of each tag

#find the most common used tags

top_tags=tags_count.most_common()
print()
#Print the top 5 tags

for tags,count in top_tags[:5]:
    print(f"{tags}:{count} times")
    
print()

#get all the quotes by einstein
def get_quotes_with_einstein():
    quotes_with_einstein = []

    response = requests.get(URL)
    soup = BS(response.text, 'html.parser')

    # All the quotes are stored in a div class quote
    for quote_div in soup.find_all('div', class_='quote'):
        # Check if Albert Einstein is mentioned in the quote
        if 'Albert Einstein' in quote_div.text:
            quote_text=quote_div.get_text(strip=True, separator='\n')
            quotes_with_einstein.append(quote_text)

    return quotes_with_einstein

quotes_with_einstein = get_quotes_with_einstein()

for idx, quote in enumerate(quotes_with_einstein, start=1):
    print(f"Quote {idx}: {quote}")
```
